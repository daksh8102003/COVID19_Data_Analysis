```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("owid-covid-data.csv")
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
df_india = df[df['location'] == 'India']
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_india['date'], df_india['new_cases'], label='New Cases')
plt.title('Daily New COVID-19 Cases in India')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
latest = df[df['date'] == df['date'].max()]
top_countries = latest.groupby('location')['total_cases'].max().sort_values(ascending=False).head(10)

top_countries.plot(kind='barh', figsize=(10,6), title='Top 10 Countries by Total COVID Cases')
plt.xlabel('Total Cases')
plt.show()
```
